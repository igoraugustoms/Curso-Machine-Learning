```python
import pandas as pd
```

```python
dados = pd.read_csv('buscas.csv', sep = ',')
dados.head()
```

```python
X = []
Y = []
for c in range(dados.shape[0]):
    r = []
    r.append(int(dados['home'][c]))
    r.append(str(dados['busca'][c]))
    r.append(int(dados['logado'][c]))
    X.append(r)
    Y.append(int(dados['comprou'][c]))
print(Y)
```

```python
x_df = dados[['home', 'busca', 'logado']]
y_df = dados['comprou']
```

```python
# variaveis dummies são variaveis que não estavam presentes nos dados que recebemos originalmente, mas que conseguimos extrair do dataset original. Exemplo: as colunas mostrando o que pesquisou... java, algoritmo, ruby
```

```python
xdummies_df = pd.get_dummies(x_df)
ydummies_df = y_df
# a funcao pd.get_dummies gera automaticamente os dummies do DataFrame em que estou trabalhando
```

```python
X = xdummies_df.values
Y = ydummies_df.values
# o método nome.values transforma o DataFrame em um array. Muito mais fácil do que eu jeito antigo com append
```

```python
porcentagem_de_treino = 0.8
porcentagem_de_teste = 0.1

tamanho_de_treino = int(porcentagem_de_treino * len(Y))
tamanho_de_teste = int(porcentagem_de_teste * len(Y))
tamanho_de_validacao = int(len(Y) - tamanho_de_treino - tamanho_de_teste)

treino_dados = X[:tamanho_de_treino]
treino_marcacoes = Y[:tamanho_de_treino]

fim_de_teste = tamanho_de_treino + tamanho_de_teste
teste_dados = X[tamanho_de_treino:fim_de_teste]
teste_marcacoes = Y[tamanho_de_treino:fim_de_teste]
                    
validacao_dados = X[fim_de_teste:]
validacao_marcacoes = Y[fim_de_teste:]
```

```python
# from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import AdaBoostClassifier
```

```python
# modelo = MultinomialNB()
modelo = AdaBoostClassifier()
```

```python
modelo.fit(treino_dados, treino_marcacoes)
```

```python
resultado = modelo.predict(teste_dados)
```

```python
diferencas = resultado - teste_marcacoes
```

```python
acertos = []
for c in range(len(diferencas)):
    if diferencas[c] == 0:
        acertos.append(diferencas[c])
```

```python
taxa_acerto = (len(acertos)/len(teste_marcacoes)) * 100
print(taxa_acerto)
```

```python
# acerto_de_um = sum(Y)
# acerto_de_zero = len(Y) - acerto_de_um
from collections import Counter
acerto_base = max(Counter(validacao_marcacoes).values())
taxa_de_acerto_base = 100.0 * acerto_base / len(validacao_marcacoes)
# taxa_de_acerto_base = 100.0 * (max(acerto_de_um, acerto_de_zero) / len(Y))

# Escolhendo a taxa de acerto base
```

```python
# prevendo a taxa de acerto no mundo real
resultado_real = modelo.predict(validacao_dados)
diferenca_real = resultado_real - validacao_marcacoes
acerto_real = []
for t in range(len(diferenca_real)):
    if diferenca_real[t] == 0:
        acerto_real.append(diferenca_real[t])
taxa_acerto_real = (len(acerto_real) / len(diferenca_real)) * 100
```

```python
print(f'A taxa de acerto base é {taxa_de_acerto_base}%')
print(f'A taxa de acerto do algoritmo é {taxa_acerto}%')
print(f'A taxa de acerto no mundo real é {taxa_acerto_real}%')
```
